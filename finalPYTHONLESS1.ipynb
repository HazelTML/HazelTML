```python
# Import the required libraries
from scipy.stats import norm  # For statistical computations
import numpy as np  # For working with arrays and mathematical functions
import matplotlib.pyplot as plt  # For plotting
import seaborn as sb  # For improving plot aesthetics

# Define the mean and standard deviation
mean = 4300  # Average acres burned
std_dev = 750  # Spread of the distribution (standard deviation)

# Define the range of interest
x1 = 2500  # Lower bound
x2 = 4200  # Upper bound

# Calculate cumulative probabilities for x1 and x2
prob_x1 = norm(loc=mean, scale=std_dev).cdf(x1)  # Probability of X <= x1
prob_x2 = norm(loc=mean, scale=std_dev).cdf(x2)  # Probability of X <= x2
prob_between = prob_x2 - prob_x1  # Probability of x1 <= X <= x2

# Print the result with percentage format
print(f"P({x1} <= X <= {x2}) = {prob_between:.2%}")  # Show result as a percentage

# Define the x range for plotting the normal distribution
data = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 1000)
# Generate 1000 evenly spaced points within ±4 standard deviations of the mean

# Calculate the probability density function (PDF) for these points
pdf = norm.pdf(data, loc=mean, scale=std_dev)  # PDF gives the height of the curve

# Define the critical values for shading
prob_x1 = x1  # Lower bound for shading
prob_x2 = x2  # Upper bound for shading

# Plot the distribution
sb.set_style('whitegrid')  # Set a clean grid style
plt.figure(figsize=(10, 6))  # Set the plot size
plt.plot(data, pdf, color='black', label='Normal Distribution')  # Plot the curve

# Shade the area between x1 and x2
plt.fill_between(data, pdf, where=(data >= prob_x1) & (data <= prob_x2),
                 color='green', alpha=0.5, label=f'Area between {x1} and {x2}')
# Highlights the region between x1 and x2 with green shading

# Add vertical lines for x1 and x2
plt.axvline(prob_x1, color='red', linestyle='--', label=f'{x1} acres')  # Vertical line at x1
plt.axvline(prob_x2, color='blue', linestyle='--', label=f'{x2} acres')  # Vertical line at x2

# Add labels, title, and legend
plt.xlabel('Acres Burned')  # X-axis label
plt.ylabel('Probability Density')  # Y-axis label
plt.title('Distribution of Acres Burned')  # Title of the plot
plt.legend()  # Display the legend

# Show the plot
plt.show()  # Render the plot
```

```python
# Import the required libraries
from scipy.stats import norm  # For statistical computations
import numpy as np  # For array manipulations and numerical calculations
import matplotlib.pyplot as plt  # For plotting
import seaborn as sb  # For improving plot aesthetics

# Define the mean and standard deviation
mean = 27989  # Average salary in PHP for first-year teachers
std_dev = 3250  # Spread of the salary distribution in PHP

# Define the salary range of interest
x1 = 20000  # Lower bound salary
x2 = 30000  # Upper bound salary

# Calculate cumulative probabilities
prob_x1 = norm(loc=mean, scale=std_dev).cdf(x1)  # Probability of X <= x1
prob_x2 = norm(loc=mean, scale=std_dev).cdf(x2)  # Probability of X <= x2
prob_between = prob_x2 - prob_x1  # Probability of x1 <= X <= x2

# Print the result as a percentage
print(f"P({x1} <= X <= {x2}) = {prob_between:.2%}")  # Output the result

# Define the x range for the plot
data = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 1000)
# Generate 1000 points spanning ±4 standard deviations from the mean

# Calculate the probability density function (PDF)
pdf = norm.pdf(data, loc=mean, scale=std_dev)  # Height of the normal curve at each point

# Define bounds for shading
prob_x1 = x1  # Lower limit of the shaded area
prob_x2 = x2  # Upper limit of the shaded area

# Plot the normal distribution
sb.set_style('whitegrid')  # Set a clean grid style for the plot
plt.figure(figsize=(10, 6))  # Set the figure size
plt.plot(data, pdf, color='black', label='Normal Distribution')  # Plot the curve

# Shade the area between x1 and x2
plt.fill_between(data, pdf, where=(data >= prob_x1) & (data <= prob_x2),
                 color='green', alpha=0.5, label=f'Area between {x1} and {x2}')
# Highlights the region between x1 and x2 with green shading

# Add vertical lines for critical values
plt.axvline(prob_x1, color='red', linestyle='--', label=f'{x1} PHP')  # Mark x1
plt.axvline(prob_x2, color='blue', linestyle='--', label=f'{x2} PHP')  # Mark x2

# Add labels, title, and legend
plt.xlabel('Salary for First-Year Teachers')  # X-axis label
plt.ylabel('Probability Density')  # Y-axis label
plt.title('Distribution of Salary for First-Year Teachers')  # Title of the plot
plt.legend()  # Display the legend

# Show the plot
plt.show()  # Render the visualization
```

```python
# Import the required library
from scipy.stats import norm

# Define the z-score
z_score = 2.30

# Calculate the probability
prob_greater_than_z = 1 - norm.cdf(z_score)

# Print the result
print(f"P(Z > {z_score}) = {prob_greater_than_z:.2%}")
```
